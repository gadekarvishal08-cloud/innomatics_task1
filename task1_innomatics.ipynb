```python
# =========================================
# 1. User Login Check
# =========================================

# Stored credentials
correct_username = "admin"
correct_password = "1234"

# Input from user
username = input("Enter username: ")
password = input("Enter password: ")

# Check login
if username == correct_username and password == correct_password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# =========================================
# 2. Pass / Fail Analyzer
# =========================================

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

# Loop through marks
for m in marks:
    if m >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("\nPass/Fail Analysis")
print("Total Pass:", pass_count)
print("Total Fail:", fail_count)
```

```python
# =========================================
# 3. Simple Data Cleaner
# =========================================

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

# Clean each name
for name in names:
    clean_name = name.strip().lower()  # remove spaces + lowercase
    cleaned_names.append(clean_name)

print("\nCleaned Names:", cleaned_names)
```

```python
# =========================================
# 4. Message Length Analyzer
# =========================================

messages = ["Hi", "Welcome to the platform", "OK"]

print("\nMessage Length Analysis")

for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' | Length: {length}")

    if length > 10:
        print("⚠️ This message is longer than 10 characters")
```

```python
# =========================================
# 5. Error Message Detector
# =========================================

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

# Count ERROR logs
for log in logs:
    if log == "ERROR":
        error_count += 1

print("\nTotal ERROR messages:", error_count)
```
