***Problem Statement: 01***

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

highest_score = max(employees.values())

best_performer = []
for name, score in employees.items():
    if score == highest_score:
        best_performer.append(name)


print("Top performer Eligible for Bonus:" ,best_performer)
print(f"(Score {highest_score})")
```

***Problem statement: 02***

```python
query = "Buy mobile phone buy phone online"

lower_query = query.lower()

word_count = {}
words = lower_query.split()
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print("Keywords searched more than once : ")

result = {}
for word, count in word_count.items():
    if count > 1:
        result[word] = count

print(result)
```

***Problem Statement: 03***

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index,reading in enumerate(sensor_readings): # We want to store both index and Readings into the valid readings,thats why we use enumerates
    if reading % 2 == 0:
        valid_readings.append((index,reading))

print("The valid readings are: ", valid_readings)
```

***Problem statement:4***

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com",
"adesh@gmail.com"
]

# Creating a Set holding only the count of each domain from email
domain_count = {}

for mails in emails:
    domain = mails.split("@")[1]

    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

#Calculating the percentage of the Mails
total = len(emails)

for domain, count in domain_count.items():
    percentage = count/total * 100

    print(f"{domain} : {int(percentage)}%")
```

***Problem Statement: 05***

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

average_sales = sum(sales)/len(sales)

value = average_sales * 0.30 # 30% of total sales
threshold = average_sales + value
# print(threshold)
for index,sale in enumerate(sales):
    if sale > threshold:
        print(f"Day {index + 1} : Sales {sale}")
```
